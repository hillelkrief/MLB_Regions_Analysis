```python
#The goal is to now aggregate our data to show the differences between specific statistics from players coming from differnt regions

```

```python
All_Batters_career_SUM = batters.groupby('playerID').agg({
    'WAR': 'sum',
    'HR': 'sum',
    'GP': 'sum',
    'H': 'sum',
    'playerID':'count',
    'region': 'first',
    'birthCountry':'first',
    'birthState':'first',
    'birthCity':'first'
})
```

```python
#Renaming GP column to represent ASG and playerID to represent seasons played
All_Batters_career_SUM = All_Batters_career_SUM.rename(columns={'GP': 'ASG'})
All_Batters_career_SUM = All_Batters_career_SUM.rename(columns={'playerID': 'seasonsPlayed'})
```

```python
All_Batters_career_SUM
```

```python
#Now, we'll do the same for pitchers
```

```python
pitchers.columns
```

```python
All_Pitchers_career_SUM = pitchers.groupby('playerID').agg({
    'WAR': 'sum',
    'ERA_plus': 'mean',
    'G_all': 'sum',
    'playerID':'count',
    'GP': 'sum',
    'region': 'first',
    'birthCountry':'first',
    'birthState':'first',
    'birthCity':'first'
})

All_Pitchers_career_SUM
```

```python
All_Pitchers_career_SUM = All_Pitchers_career_SUM.rename(columns={'playerID': 'seasonsPlayed'})

All_Pitchers_career_SUM = All_Pitchers_career_SUM.rename(columns={'GP': 'ASG'})
All_Pitchers_career_SUM
```
