# Maps!

```python

data_to_plot_b = war_batters[['birthCountry','WAR_mean']]
data_to_plot_b.head()
```

```python

data_to_plot_p = war_pitchers[['birthCountry','WAR_mean']]
data_to_plot_p.head()
```

# Batter Map

```python
data_to_plot_b.loc[data_to_plot_b['birthCountry'] == 'USA', 'birthCountry'] = 'United States of America'
```

```python

# Setup a folium map at a high-level zoom
map = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = country_geo, 
    data = data_to_plot_b,
    columns = ['birthCountry', 'WAR_mean'],
    key_on = 'feature.properties.admin', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "WAR").add_to(map)
folium.LayerControl().add_to(map)

map
```

```python
map.save('batter_war_data.html')
```

Each country is different, but we clearly can see a strength in specifc countries from one part of the world! Even within latin America we see an area that has clearly been more succeessful once reahting the major leauges
