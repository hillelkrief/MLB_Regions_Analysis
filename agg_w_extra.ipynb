```python
#The goal is to now aggregate our data to show the differences between specific statistics from players coming from differnt regions

```

```python
All_Batters_career_SUM = batters.groupby('playerID').agg({
    'WAR': 'sum',
    'HR': 'sum',
    'GP': 'sum',
    'H': 'sum',
    'playerID':'count',
    'region': 'first'
})
```

```python
#Renaming GP column to represent ASG and playerID to represent seasons played
All_Batters_career_SUM = All_Batters_career_SUM.rename(columns={'GP': 'ASG'})
All_Batters_career_SUM = All_Batters_career_SUM.rename(columns={'playerID': 'seasonsPlayed'})
```

```python
All_Batters_career_SUM
```

```python
#Batter WAR average
BatterWarAvg=All_Batters_career_SUM.groupby('region').agg({'WAR': ['mean']})
BatterWarAvg
```

```python
#Batter WAR average for WAR above 0
BatterWarAvgAdj=All_Batters_career_SUM[All_Batters_career_SUM['WAR'] > 0].groupby('region').agg({'WAR': ['mean']})
BatterWarAvgAdj
```

```python
#HR average for HR's above 0
HRAvgAdj=All_Batters_career_SUM[All_Batters_career_SUM['HR'] > 0].groupby('region').agg({'HR': ['mean']})
HRAvgAdj
```

```python
#HR average
HRAvg=All_Batters_career_SUM.groupby('region').agg({'HR': ['mean']})
HRAvg
```

```python
#allstar average
ASGBatters=All_Batters_career_SUM.groupby('region').agg({'ASG': ['mean']})
ASGBatters
```

```python
#allstar average for all stars above 0
ASGBattersAdj=All_Batters_career_SUM[All_Batters_career_SUM['ASG'] > 0].groupby('region').agg({'ASG': ['mean']})
ASGBattersAdj
```

```python
#Season Played average
SPAverage=All_Batters_career_SUM.groupby('region').agg({'seasonsPlayed': ['mean']})
SPAverage
```

```python
#Season Played average for seaons above 1
SPAverageAdj=All_Batters_career_SUM[All_Batters_career_SUM['seasonsPlayed'] > 0].groupby('region').agg({'seasonsPlayed': ['mean']})
SPAverageAdj
```

```python
All_Pitchers_career_SUM = pitchers.groupby('playerID').agg({
    'WAR': 'sum',
    'ERA_plus': 'mean',
    'G_all': 'sum',
    'playerID':'count',
    'GP': 'sum',
    'region': 'first'
})

All_Pitchers_career_SUM
```
